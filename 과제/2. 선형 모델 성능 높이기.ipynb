# California 데이터 선형 회귀분석

## 필요한 라이브러리 불러오기

```
import multiprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['seaborn-whitegrid'])
```

```
from sklearn.datasets import fetch_california_housing
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, OrthogonalMatchingPursuit, ElasticNet
```

## California 데이터 분석하기

```
california = fetch_california_housing()

print(california.keys())
print(california.DESCR)
```

## California 최적 파라미터 찾기

```
X_train, X_test, y_train, y_test = train_test_split(california.data, california.target, test_size=0.2)
```

```
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### Lasso

```
estimator = Lasso()
TF = [True, False]
alpha = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10] 
tol = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10] 
param_grid = {'alpha':alpha,
              'copy_X':TF,
              'fit_intercept':TF,
              'normalize':TF,
              'positive':TF,
              'precompute':TF,
              'random_state':TF,
              'tol':tol,
              'warm_start':TF}

gs = GridSearchCV(
    estimator=estimator,
    param_grid=param_grid,
    n_jobs=multiprocessing.cpu_count(),
    cv=10,
    verbose=True
)
result = gs.fit(X_train, y_train)

print("최적의 점수: {}".format(result.best_score_))
print("최적의 파라미터: {}".format(result.best_params_))
print(gs.best_estimator_)
pd.DataFrame(result.cv_results_);
```

### Ridge

```
estimator = Ridge()
TF = [True, False]
alpha = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1] 
tol = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1] 
param_grid = {'alpha':alpha,
              'copy_X':TF,
              'fit_intercept':TF,
              'normalize':TF,
              'random_state':TF,
              'tol':tol,}

gs = GridSearchCV(
    estimator=estimator,
    param_grid=param_grid,
    n_jobs=multiprocessing.cpu_count(),
    cv=10,
    verbose=True
)
result = gs.fit(X_train, y_train)

print("최적의 점수: {}".format(result.best_score_))
print("최적의 파라미터: {}".format(result.best_params_))
print(gs.best_estimator_)
pd.DataFrame(result.cv_results_);
```

### ElasticNet

```
estimator = ElasticNet()
TF = [True, False]
alpha = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1] 
tol = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1] 
param_grid = {'alpha':alpha,
              'copy_X':TF,
              'fit_intercept':TF,
              'normalize':TF,
              'random_state':TF,
              'tol':tol,}
              

gs = GridSearchCV(
    estimator=estimator,
    param_grid=param_grid,
    n_jobs=multiprocessing.cpu_count(),
    cv=10,
    verbose=True
)
result = gs.fit(X_train, y_train)

print("최적의 점수: {}".format(result.best_score_))
print("최적의 파라미터: {}".format(result.best_params_))
print(gs.best_estimator_)
pd.DataFrame(result.cv_results_);
```

### OrthogonalMatching

```
estimator = OrthogonalMatchingPursuit(n_nonzero_coefs=5)
TF = [True, False]
tol = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1] 
param_grid = {'fit_intercept':TF,
              'normalize':TF,
              'tol':tol}


gs = GridSearchCV(
    estimator=estimator,
    param_grid=param_grid,
    n_jobs=multiprocessing.cpu_count(),
    cv=10,
    verbose=True
)
result = gs.fit(X_train, y_train)

print("최적의 점수: {}".format(result.best_score_))
print("최적의 파라미터: {}".format(result.best_params_))
print(gs.best_estimator_)
pd.DataFrame(result.cv_results_);
```

### Linear Regression

```
estimator = LinearRegression()
TF = [True, False]
param_grid = {'copy_X':TF,
              'fit_intercept':TF,
              'normalize':TF}

gs = GridSearchCV(
    estimator=estimator,
    param_grid=param_grid,
    n_jobs=multiprocessing.cpu_count(),
    cv=10,
    verbose=True
)
result = gs.fit(X_train, y_train)

print("최적의 점수: {}".format(result.best_score_))
print("최적의 파라미터: {}".format(result.best_params_))
print(gs.best_estimator_)
pd.DataFrame(result.cv_results_);
```
